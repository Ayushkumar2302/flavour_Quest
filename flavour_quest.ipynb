                                                         Data Cleaning

```python
# Strip whitespace & lowercase categorical columns
for col in ['diet', 'course', 'flavor_profile', 'state']:
    df[col] = df[col].astype(str).str.strip().str.lower()
```

```python
# Replace blank or 'nan' flavor profiles with 'unknown'
df['flavor_profile'] = df['flavor_profile'].replace(['nan', 'none', '', 'unknown'], np.nan)
df['flavor_profile'].fillna('unknown', inplace=True)
```

```python
# Clean ingredients into a list
df['ingredients_cleaned'] = df['ingredients'].dropna().apply(
    lambda x: [i.strip().lower() for i in x.split(',')]
)
```

```python
df.drop_duplicates(subset=['name', 'ingredients'], inplace=True)

df.head()
```

```python
all_ingredients = [ingredient for sublist in df['ingredients_cleaned'].dropna() for ingredient in sublist]
ingredient_counts = Counter(all_ingredients)
```

```python
top_ingredients = pd.DataFrame(ingredient_counts.most_common(20), columns=['Ingredient', 'Count'])
```

```python
# ----------------------------------------
# 📌 STEP 4: Data Cleaning
# ----------------------------------------

# Strip whitespace & lowercase categorical columns
for col in ['diet', 'course', 'flavor_profile', 'state']:
    df[col] = df[col].astype(str).str.strip().str.lower()

# Replace blank or 'nan' flavor profiles with 'unknown'
df['flavor_profile'] = df['flavor_profile'].replace(['nan', 'none', '', 'unknown'], np.nan)
df['flavor_profile'].fillna('unknown', inplace=True)

# Clean ingredients into a list
df['ingredients_cleaned'] = df['ingredients'].dropna().apply(
    lambda x: [i.strip().lower() for i in x.split(',')]
)

# Drop duplicates if any
df.drop_duplicates(subset=['name', 'ingredients'], inplace=True)

df.head()
```

```python
plt.figure(figsize=(12,6))
```

```python
sns.barplot(x='Count', y='Ingredient', data=top_ingredients, palette='Spectral')
plt.title('Top 20 Most Common Ingredients in Indian Dishes')
plt.xlabel('Frequency')
plt.ylabel('Ingredient')
plt.tight_layout()
plt.show()
```

```python
# ----------------------------------------
# 📌 STEP 5: Ingredient Frequency Analysis
# ----------------------------------------
# Flatten the list of all ingredients
all_ingredients = [ingredient for sublist in df['ingredients_cleaned'].dropna() for ingredient in sublist]
ingredient_counts = Counter(all_ingredients)

# Top 20 Ingredients
top_ingredients = pd.DataFrame(ingredient_counts.most_common(20), columns=['Ingredient', 'Count'])

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x='Count', y='Ingredient', data=top_ingredients, palette='Spectral')
plt.title('Top 20 Most Common Ingredients in Indian Dishes')
plt.xlabel('Frequency')
plt.ylabel('Ingredient')
plt.tight_layout()
plt.show()
```

                                                                   **Flavor Profile Distribution**

```python
# ----------------------------------------
# 📌 STEP 6: Flavor Profile Distribution
# ----------------------------------------
flavor_counts = df['flavor_profile'].value_counts()

plt.figure(figsize=(8,5))
flavor_counts.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'), shadow=True)
plt.title('Flavor Profile Distribution')
plt.ylabel('')
plt.show()
```

                                               **Diet Type Distribution**

```python
# ----------------------------------------
# 📌 STEP 7: Diet Type Distribution
# ----------------------------------------
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='diet', palette='Set2')
plt.title('Diet Type Distribution (Veg vs Non-Veg)')
plt.xlabel('Diet')
plt.ylabel('Number of Dishes')
plt.show()
```

```python
# ----------------------------------------
# 📌 STEP 8: Dishes by State
# ----------------------------------------
top_states = df['state'].value_counts().head(15)

plt.figure(figsize=(12,6))
sns.barplot(x=top_states.values, y=top_states.index, palette='mako')
plt.title('Top 15 States with Most Dishes')
plt.xlabel('Number of Dishes')
plt.ylabel('State')
plt.show()
```

```python
# ----------------------------------------
# 📌 STEP 9: Course Distribution
# ----------------------------------------
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='course', order=df['course'].value_counts().index, palette='Set1')
plt.title('Course Type Distribution')
plt.xlabel('Course (Main/Snack/Dessert)')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ----------------------------------------
# 📌 STEP 10: Word Cloud of Ingredients
# ----------------------------------------
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(all_ingredients))

plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Ingredients Word Cloud')
plt.show()
```
